```python
import pandas as pd
import os
from sklearn.utils import shuffle
```

## Ratings

```python
df_ratings = pd.read_csv(os.path.join("data", "ml-latest-small", "ratings.csv"))
df_ratings.head()
```

```python
users = df_ratings.groupby("userId").apply(lambda df: df.shape[0]).sort_values(ascending=False).cumsum()
users_id = users[users<70585].index
df_ratings = df_ratings[df_ratings.userId.isin(users_id)]
```

```python
df_ratings = df_ratings[df_ratings["userId"]!=25]
df_ratings = df_ratings[df_ratings["userId"]!=30]
```

```python
df_new_user_1 = pd.DataFrame({"userId":[25, 25, 25, 25, 25, 25, 25, 25, 25, 25, 25, 25, 25, 25], "movieId": [260, 1196, 1210, 2628,33493, 179819, 1270, 924, 1097, 1726, 1580, 1240, 5, 357], "rating": [5, 5, 5, 5, 5, 5, 5, 5, 4.5, 4.5, 1, 5, 5, 5]})
```

```python
df_new_user_2 = pd.DataFrame({"userId":[30, 30, 30, 30, 30, 30, 30, 30, 30, 30, 30, 30], "movieId": [260, 1196, 1210, 2628, 1371, 1374, 1270, 2571,  1240, 338, 188797, 11], "rating": [5, 5, 4.5, 4.5, 5, 5, 4.5, 5, 4, 4, 2, 5]})
```

```python
df_ratings = pd.concat([df_ratings[["userId", "movieId", "rating"]], df_new_user_1, df_new_user_2])
df_ratings.head()
```

```python
users_id = df_ratings.userId.index
```

```python
df_ratings = shuffle(df_ratings.set_index("userId"))
```

```python
df_ratings
```

```python
df_ratings.to_csv("data/ratings.csv")
```

## Movies

```python
df_movies = pd.read_csv(os.path.join("data", "ml-latest-small", "movies.csv"))
```

```python
def extract_year(value):
    value = value.strip()
    try:
        year = value[-5:-1]
        return int(year)
    except:
        return None
```

```python
def clean_title(value):
    value = value.strip()
    try:
        year = value[-5:-1]
        year =  int(year)
        return value[:-6].strip()
    except:
        return value
```

```python
df_movies["year"] = df_movies.title.apply(extract_year)
```

```python
df_movies["title"] = df_movies.title.apply(clean_title)
```

```python
df_movies = df_movies.set_index("movieId")
```

```python
df_movies.to_csv("data/movies.csv")
```
